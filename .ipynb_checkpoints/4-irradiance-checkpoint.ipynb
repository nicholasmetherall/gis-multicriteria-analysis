```python
import matplotlib.pyplot as plt

# Replace 'your_file.tif' with your file path
tif_path = 'SI_DNI.tif'

# Open the raster file
with rasterio.open(tif_path) as src:
    # Read the first band
    band1 = src.read(1)
    # Get extent for correct plotting
    extent = [src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top]

# Plot
plt.figure(figsize=(8, 6))
plt.imshow(band1, cmap='viridis', extent=extent)
plt.colorbar(label='Value')
plt.title('Raster Plot Viridis')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np

with rasterio.open('SI_DNI.tif') as src:
    img = src.read([1, 2, 3])  # Read R, G, B bands

# Transpose from (bands, rows, cols) to (rows, cols, bands)
img = np.transpose(img, (1, 2, 0))

plt.figure(figsize=(8, 8))
plt.imshow(img)
plt.axis('off')
plt.title('Original Colors (RGB)')
plt.show()
```
